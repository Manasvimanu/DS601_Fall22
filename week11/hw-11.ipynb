```python
#IMF ESTIMATIONS for 2020 through 2027
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_past_and_projected_GDP_(nominal)'
```

```python
lag1 = pd.read_html(url, header = 0)
```

```python
len(lag1)
```

```python
lag = lag1[5]
lag.head(50)
```

```python
lag.rename(columns={'Country (or dependent territory)': 'COUNTRY'}, inplace=True)
lag.head(50)
```

```python
DF = pd.merge(lag,data,on = "COUNTRY", how = 'left')
DF.head(8)
```

```python
DF = DF.replace(np.nan, 0)
```

```python
DF.head(8)
```

```python
#calculating the average value which increase in terms of percentage.

DF['average'] = ((DF['2027']-DF['2022'])*100)/DF['2022']
```

```python
DF
```

```python
#Sorted by this column in descending order.

DF.sort_values(by=['average'], ascending = False)
```
