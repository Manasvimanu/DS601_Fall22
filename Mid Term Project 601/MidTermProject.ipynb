## Visualization

```python
# Displaying the histogram plot 
plt.hist(x = 'Total Number of Projected/Retained Jobs', data = df,  color = "orange")
```

Visualizing the data using Histogram which shows total no of projected/retained jobs 

```python
# Displaying the count plot
plt.figure(figsize=(20,10))
sns.countplot(df['Fiscal Year'], hue = df['Total Number of Employees Retained'])
plt.show()
```

Visualizing the data using count plot comparing the fiscal year with Total Number of Employees Retained

```python
# Binary Features
plt.figure(figsize=(22, 6))


plt.subplot(131)
sns.countplot(df['Total Number of Projected/Retained Jobs'])
plt.xlabel('Retained jobs')
plt.ylabel('Count')


plt.subplot(132)
sns.countplot(df['Total Projected Number of Jobs'])
plt.xlabel('Projected Number of Jobs')
plt.ylabel('Count')


plt.subplot(133)
sns.countplot(df['Total Number of Employees Retained'])
plt.xlabel('Employees Retained')
plt.ylabel('Count')
```

#### It represents the no of observations for each categorical variable(Retained jobs, Projected Number of Jobs and Employees Retained) and displays the result as bar chart

```python
sns.displot(data=df,kind="ecdf",x="Fiscal Year",hue="Total Number of Projected/Retained Jobs",aspect=1.5)
plt.xlabel('Fiancial Year')
plt.ylabel('Retained Jobs')


sns.displot(data=df,kind="ecdf",x="Fiscal Year",hue="Total Projected Number of Jobs",aspect=1.5)
plt.xlabel('Fiancial Year')
plt.ylabel('Projected Number of Jobs')


sns.displot(data=df,kind="ecdf",x="Fiscal Year",hue="Total Number of Employees Retained",aspect=1.5)
plt.xlabel('Fiancial Year')
plt.ylabel('Employees Retained')
```

#### The above three graphs represent the Number of Jobs Retained, Projected number of jobs and Employee retained in the financial year

```python
plot = sns.displot(df, x="Ward", y="Total Number of Projected/Retained Jobs", cbar = True)
plot.ax.tick_params(labelrotation = 45)
#plot.ax.set_title('Distribution between Counties and License types')
```

#### Visualizing the data using displot comparing Ward in x-axis and Total Number of Projected/Retained Jobs in y-axis

```python
plt.scatter(data = df, x="Ward", y="Total Number of Employees Retained")
plt.xlabel("Ward")
plt.ylabel("Total Number of Employees Retained")
```

#### Visualizing the data using scatterplot comparing ward and no of employees retained

```python
plot = sns.displot(df, x="Ward", y="Total Projected Number of Jobs", cbar = True)
plot.ax.tick_params(labelrotation = 45)
```

#### Visualizing the data using scatterplot comparing ward and projected number of jobs

```python
plot = sns.displot(df, x="Fiscal Year", y="Official Amount", height = 10)
plot.ax.tick_params()
```
