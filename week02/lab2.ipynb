## Lab Part1

Please choose appropriate graph and visualize the followings with correct/appropriate labels

* Height only
* Points only
* Height vs Points

```python
import pandas as pd

nba_players = pd.read_csv('https://raw.githubusercontent.com/msaricaumbc/DS_data/master/all_seasons.csv', index_col = 0)

height = nba_players['player_height']
points = nba_players['pts']
```

```python
import matplotlib.pyplot as plt
lst_height=[]
height_value=[]

for j in range(len(height)):
    lst_height.append(height[j])
    
for j in range(len(height)):
    height_value.append(j)
```

```python
#plt is for players height's

#Scatter plt is for height of players
plt.figure(figsize=(5,6))
plt.title("Height visualisation")
plt.scatter(lst_height,height_value)
plt.xlabel('Height_of_players')
plt.ylabel("Player_No")
plt.show()

#Hist is for height of player
plt.figure(figsize=(5,6))
plt.title("Height visualisation")
plt.hist(lst_height)
plt.xlabel("Player_Height")
plt.ylabel("No_of_players")
plt.grid(True)
plt.show()
```

```python
points_lst = []
player_no_lst = []

for i in range(len(points)):
    points_lst.append(points[i])
for i in range(len(points)):
    player_no_lst.append(i)

#plt is for points

plt.title("Points visualisation")
plt.xlabel("Players_No")
plt.ylabel("Points_gained")
plt.bar(player_no_lst,points_lst)
plt.show()
```

```python
#plt is for height vs  points

height_lst=[]
for i in range(len(height)):
    height_lst.append(i)
plt.scatter(lst_height,points_lst)
plt.title("Players height vs Points")
plt.ylabel('Points')
plt.xlabel('Player_height')
plt.draw()
```

## Lab Part2
Please visualize year vs score with correct/appropriate labels

```python
bowling_data = pd.read_csv("https://raw.githubusercontent.com/msaricaumbc/DS_data/master/bowling_stats.csv",
                            header=None,
                            names=['year','city','state','count1','count2'])
def merge_columns(row):
    if pd.isna(row['count2']):
        return row['count1']
    else:
        return row['count1']*1000+row['count2']
    
bowling_data['total']=bowling_data.apply(merge_columns,axis=1)

year = bowling_data['year']
score = bowling_data['total']
Year=[]
Score=[]
for i in range(len(year)):
    Year.append(year[i])
for j in range(len(score)):
    Score.append(score[j])

#plt is for yearvsscore
plt.figure(figsize=(20,4))
plt.title('Year_vs_Score')
plt.xlabel("Year")
plt.ylabel("Score")
plt.xticks(rotation = 90)
plt.scatter(Year,Score)
plt.show()
```
