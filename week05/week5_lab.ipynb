```python
import seaborn as sns
import pandas as pd

diamonds = sns.load_dataset("diamonds")

diamonds.head()
```

#### 1.

- Plot the scatter plot of price and carat. Add clarity into the plot as hue. 

- Do you see any pattern? Is it parallel to your expectation?

```python
a = diamonds.price
b = diamonds.carat
H = diamonds.clarity
plt = sns.scatterplot(x= a, y = b, hue = H, alpha = 0.5 )
```

In the above graph, as the price increases the carat and clarity also increases

# 2.

- Consider `clarity` and `color` is there any pattern in their heatmap?

```python
Plt= sns.displot(data=diamonds, kind="hist", x="clarity",y="color", aspect = 2.0)
```

In the above map, where ever the color is dark, the clarity and color also increases
